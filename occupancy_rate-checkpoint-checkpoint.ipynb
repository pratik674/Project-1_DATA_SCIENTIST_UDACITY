### Question: What is the occupancy rate for the next 30days? what is the hosts revenue in different parts of Seattle?

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

calendar_df = pd.read_csv('C:/Users/prati/OneDrive/Desktop/calendar.csv')
listing_df = pd.read_csv('C:/Users/prati/OneDrive/Desktop/listings.csv')
reviews_df = pd.read_csv('C:/Users/prati/OneDrive/Desktop/reviews.csv')
listing_df.shape
```

Data cleaning. 
For the columns pertaining to occupancy rate, find out if there are any missing values. IF so, then remove these values before proceeding. Calculate occupancy rate.

```python
# no of missing values in column availability_30
listing_df['availability_30'].isnull().sum()
```

```python
listing_df['perc_occ_30d'] = 1- listing_df['availability_30']/30
df_temp2 = listing_df['perc_occ_30d']
df_temp2
```

Create a new data frame merging the dataframe extracted above with the dataframe of columns pertaining to neighbourhood and property type. Plot the variation of occupancy rate with each neighbourhood

```python
df_temp1 = listing_df[['neighbourhood','neighbourhood_group_cleansed','property_type']]
df_concat = pd.concat([df_temp1,df_temp2], axis =1, sort=False)


df = df_concat.groupby(['neighbourhood_group_cleansed','neighbourhood']).mean()['perc_occ_30d'].reset_index(name = 'avg_occ_rate_30d')
df.shape

df1= df.sort_values(by='avg_occ_rate_30d', ascending=False)
print(df1.shape)

sns.catplot(data = df1, x = 'neighbourhood_group_cleansed', y = 'avg_occ_rate_30d', hue='neighbourhood', kind ='bar', dodge=False, palette='rocket_r', height = 5, aspect = 5)
plt.title('Occupancy rate of top most neighbourhood ')
plt.show()




```
